```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from collections import defaultdict
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import lightgbm as lgb
```

```python
dataframe = pd.read_csv(path)
dataframe.head()
```

```python
dataframe.info()
```

```python
print("shape of dataframe : ",dataframe.shape)
```

```python
dataframe = dataframe.drop(columns=['customer_id','name','gender','no_of_children','no_of_days_employed','migrant_worker'])
dataframe = dataframe.dropna()
print("shape of dataframe : ",dataframe.shape)
```

```python
dataframe = dataframe.drop_duplicates()
print("shape of dataframe : ",dataframe.shape)
```

```python
dataframe['credit_card_default'].value_counts()
```

```python
X = dataframe.drop(columns=['credit_card_default'])
y = dataframe.drop(dataframe.iloc[:, :dataframe.shape[1]-1],axis = 1)
print("X shape : ",X.shape)
print("y shape : ",y.shape)
```

```python
d = defaultdict(LabelEncoder)

def encode(col):
    if col.dtype == 'object':
        return d[str(col)].fit_transform(col)
    else:
        return col 
    
X = X.apply(encode)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("X_train shape : ",X_train.shape)
print("y_train shape : ",y_train.shape)

print("X_test shape : ",X_test.shape)
print("y_test shape : ",y_test.shape)
```

```python
clf = CatBoostClassifier(
    iterations=500,
    max_depth=6,
    random_seed=42,
    learning_rate=0.01,
    custom_loss=['AUC', 'Accuracy'])

clf.fit(X_train, y_train,eval_set=(X_test, y_test),verbose=False)

print(f'Train accuracy: {clf.score(X_train,y_train)}')
print(f'Test accuracy: {clf.score(X_test,y_test)}')
```

```python
model = RandomForestClassifier(n_estimators=100
                               ,max_depth=10,random_state=42)
model.fit(X_train,y_train)

print(f'Train accuracy: {accuracy_score(y_train,model.predict(X_train))}')
print(f'Test accuracy:  {accuracy_score(y_test,model.predict(X_test))}')
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf1 = AdaBoostClassifier(n_estimators=100, learning_rate=0.01, random_state=15)
clf1.fit(X_train, y_train)

print(f'Train accuracy: {accuracy_score(y_train,clf1.predict(X_train))}')
print(f'Test accuracy:  {accuracy_score(y_test,clf1.predict(X_test))}')
```

```python
from sklearn.ensemble import BaggingClassifier

bagg = BaggingClassifier(max_samples=0.05)
bagg.fit(X_train,y_train)

print(f'Train accuracy: {accuracy_score(y_train,bagg.predict(X_train))}')
print(f'Test accuracy:  {accuracy_score(y_test,bagg.predict(X_test))}')
```

```python
from xgboost import XGBClassifier as xgb
# fit model no training data
xgb1 = xgb(max_depth = 6,
                                 learning_rate = 0.01,
                                 n_estimators = 10000,
                                 objective = 'binary:logistic',
                                 tree_method = 'gpu_hist',
                                 booster = 'gbtree',
                                 seed = 23,
                                 min_child_weight = 35,
                                 subsample = 0.7,
                                 alpha = 0.25,
                                 n_jobs = -1
                                 )
xgb1.fit(X_train, y_train)
print(f'Train accuracy: {accuracy_score(y_train,xgb1.predict(X_train))}')
print(f'Test accuracy:  {accuracy_score(y_test,xgb1.predict(X_test))}')
```

```python
l1 = lgb.LGBMClassifier(objective='binary',
                                   n_estimators=1000,
                                   learning_rate=0.02,
                                   num_leaves=34,
                                   colsample_bytree=0.9497036,
                                   subsample=0.8715623,
                                   max_depth=8,
                                   reg_alpha=0.041545473,
                                   reg_lambda=0.0735294,
                                   min_split_gain=0.0222415,
                                   min_child_weight=39.3259775,
                                   silent=-1,
                                   random_state = 42)

l1.fit(X_train, y_train.values.ravel(),
      
      eval_set = [(X_train, y_train), (X_test, y_test)], 
                       eval_metric ='auc',
                       early_stopping_rounds = 800)

print(f'Train accuracy: {accuracy_score(y_train,l1.predict(X_train))}')
print(f'Test accuracy:  {accuracy_score(y_test,l1.predict(X_test))}')
```

```python
from sklearn.linear_model import LogisticRegression

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

lr = LogisticRegression(C=100000.0, random_state=0)
lr.fit(X_train_std, y_train)

print(f'Train accuracy: {lr.score(X_train_std,y_train)}')
print(f'Test accuracy: {lr.score(X_test_std,y_test)}')
```
